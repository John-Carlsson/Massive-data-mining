# Coding of strings

```python
# standard way
B_coef = 2**30

def str_coding(s, B):
    base = 8 # base ASCII code
    print("the string \"%s\" has length = %d\n" % (s, len(s)))
    code_sum = 0
    group = 1
    while B >= code_sum:
        print("the code %d is less than %d, let's try with grouping one more character" % (code_sum, B))
        code_sum = 0
        for i in range(0, len(s), group):
            code_sum_sub = 0
            group_pos = 0
            for c in s[i:i + group]: #evaluation with notational positional
                code_sum_sub += ord(c) * ((2**base) ** group_pos)
                group_pos += 1
            print(s[i:i + group], code_sum_sub)
            code_sum += code_sum_sub
        print("the string code is greater than or equal than %d: %d" % (B, code_sum))  # debug info. Delete me
        group += 1

    return code_sum



words = "Nel mezzo del cammin di nostra vita"
print("The final string code is %d" % str_coding(words, B_coef))
```

```python
# Example with an external text file
f_in = open("sample.txt")
words = f_in.read()
f_in.close()
print(str_coding(words, B_coef))
```
